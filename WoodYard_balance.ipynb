```python
# @title Служебная вкладка
#импорт библиотек
import pandas as pd

#Определение классов
class WoodYard:
  def __init__(self, chip_consumption, wood_type, regime):
    self.chip_consumption = chip_consumption
    self.wood_type = wood_type
    self.regime = regime
  def wood_param():
     return {'Береза':{'Выход':[85,'%'],
                       'Сухость':[92.5,'%'],
                       'Средняя расчетная плотность древесины':[0.5,'а.с.т/пл.м3'],
                       'Средняя расчетная плотность коры':[0.43,'а.с.т/пл.м3'],
                       'Естественная влажность древесины':[50,'%'],
                       'Естественная влажность коры':[50,'%'],
                       'Потери щепы на станции сортирования':[2,'%'],
                       'Потери щепы при хранении на складе щепы':[0.5,'%'],
                       'Потери древесины при окорке':[3,'%'],
                       'Потери древесины при хранении на складе круглых лесоматериалов':[0.5,'%'],
                       'Содержание коры на круглом лесоматериале':[12,'%'],
                       'Содержание коры в щепе на варку':[0.5,'%']
                       },
            'Хвойные':{'Выход':[85,'%'],
                       'Сухость':[92.5,'%'],
                       'Средняя расчетная плотность древесины':[0.39,'а.с.т/пл.м3'],
                       'Средняя расчетная плотность коры':[0.43,'а.с.т/пл.м3'],
                       'Естественная влажность древесины':[50,'%'],
                       'Естественная влажность коры':[50,'%'],
                       'Потери щепы на станции сортирования':[2,'%'],
                       'Потери щепы при хранении на складе щепы':[0.5,'%'],
                       'Потери древесины при окорке':[3,'%'],
                       'Потери древесины при хранении на складе круглых лесоматериалов':[0.5,'%'],
                       'Содержание коры на круглом лесоматериале':[12,'%'],
                       'Содержание коры в щепе на варку':[0.5,'%']
                       },
            'Лиственница':{'Выход':[85,'%'],
                       'Сухость':[92.5,'%'],
                       'Средняя расчетная плотность древесины':[0.57,'а.с.т/пл.м3'],
                       'Средняя расчетная плотность коры':[0.43,'а.с.т/пл.м3'],
                       'Естественная влажность древесины':[50,'%'],
                       'Естественная влажность коры':[50,'%'],
                       'Потери щепы на станции сортирования':[2,'%'],
                       'Потери щепы при хранении на складе щепы':[0.5,'%'],
                       'Потери древесины при окорке':[3,'%'],
                       'Потери древесины при хранении на складе круглых лесоматериалов':[0.5,'%'],
                       'Содержание коры на круглом лесоматериале':[12,'%'],
                       'Содержание коры в щепе на варку':[0.5,'%']
                       }
             }
  def utility_consumption():
     return {
            'Удельный расход св. воды (мех.очищенной) на ДПП':[0.4,'м3/пл.м3'],
            'Удельный расход пара на ДПП':[92.5,'т/пл.м3'],
            'Удельный расход электроэнергии на ДПЦ':[10,'кВт ч/пл.м3'],
            'Удельный расход электроэнергии на прочие объекты':[7,'кВт ч/пл.м3'],
            'Удельный расход сжатого воздуха на ДПЦ':[0.001,'нм3/пл.м3'],
            }
  def wood_balance(self):
    cat = WoodYard.wood_param()[wood_type]
    a = self.chip_consumption
    b = 0.01*cat['Содержание коры в щепе на варку'][0]
    c = 1/(1-0.01*cat['Содержание коры на круглом лесоматериале'][0])-1
    sort_loss = 1-0.01*cat['Потери щепы на станции сортирования'][0]
    stor_loss = 1-0.01*cat['Потери щепы при хранении на складе щепы'][0]
    debark_loss = 1-0.01*cat['Потери древесины при окорке'][0]
    log_loss = 1-0.01*cat['Потери древесины при хранении на складе круглых лесоматериалов'][0]
    chip_dens = cat['Средняя расчетная плотность древесины'][0]
    bark_dens = cat['Средняя расчетная плотность коры'][0]
    res = {}
    c1 = a*(1-b)
    res['Щепа на производство РЦ'] = [c1, 'а.с.т/год']
    c1 = c1/sort_loss
    res['Щепа на сортирование'] = [c1, 'а.с.т/год']
    c1 = c1/stor_loss
    res['Щепа на склад хранения'] = [c1, 'а.с.т/год']
    c1 = c1/debark_loss
    res['Древесина в ДПЦ'] = [c1, 'а.с.т/год']
    c1 = c1/log_loss
    res['Годовая потребность древесины'] = [c1, 'а.с.т/год']
    b1 = a*b
    res['Кора(0,5%) на производство РЦ'] = [b1, 'а.с.т/год']
    b1 = b1/sort_loss
    res['Кора(0,5%) на сортирование'] = [b1, 'а.с.т/год']
    b1 = b1/stor_loss
    res['Кора(0,5%) на склад хранения'] = [b1, 'а.с.т/год']
    b1 = c1*c*log_loss*bark_dens/chip_dens
    res['Кора(12%)  в ДПЦ'] = [b1, 'а.с.т/год']
    b1 = b1/log_loss
    res['Годовая потребность коры'] = [b1, 'а.с.т/год']
    return res

  def comp_air_consumption(self):
    cat = WoodYard.wood_param()[wood_type]
    cons = WoodYard.utility_consumption()
    wood_balance_res = WoodYard.wood_balance(self)
    chip_dens = cat['Средняя расчетная плотность древесины'][0]
    bark_dens = cat['Средняя расчетная плотность коры'][0]
    a = wood_balance_res['Годовая потребность древесины'][0]/bark_dens + wood_balance_res['Годовая потребность коры'][0]/chip_dens
    comp_air_year = a*cons['Удельный расход сжатого воздуха на ДПЦ'][0]
    comp_air_day = comp_air_year/self.regime
    comp_air_hour = comp_air_day/24
    return {'Расход сжатого воздуха, год': [comp_air_year,'нм3/год'],
            'Расход сжатого воздуха, сутки': [comp_air_day,'нм3/сут'],
            'Расход сжатого воздуха, час': [comp_air_hour,'нм3/час']}

```
